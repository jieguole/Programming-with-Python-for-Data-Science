```python
from sklearn.decomposition import PCA
import pandas as pd
```

```python
df = pd.read_csv("Datasets/kidney_disease.csv")
df = df.drop(labels=['id', 'rbc', 'pc', 'pcc', 'ba', 'cad', 'appet', 'pe', 'ane', 'pcv', 'wc', 'rc', 'htn', 'dm', 'classification'], axis = 1)
df= df.fillna(0)
df
```

```python
pca = PCA(n_components=2, svd_solver='randomized')
pca.fit(df)
PCA(copy=True, n_components=2, whiten=False)

T = pca.transform(df)

df
```
