```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
matplotlib.style.use('ggplot') # Look Pretty
```

```python
student_dataset = pd.read_csv("students.data", index_col=0)
student_dataset
```

```python
my_series = student_dataset.G3
my_dataframe = student_dataset[['G3', 'G2', 'G1']] 
```

```python
my_series.plot.hist(alpha=0.5,bins=15)
my_dataframe.plot.hist(alpha=0.5)
plt.show()
```

```python
student_dataset.plot.scatter(x='G1', y='G2')
plt.suptitle('G1 vs G2')
plt.show()
```

```python
student_dataset.plot.scatter(x='G1', y='G3')
plt.suptitle('G1 vs G3')
plt.show()
```

```python
student_dataset.plot.scatter(x='G2', y='G3')
plt.suptitle('G2 vs G3')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel('Final Grade')
ax.set_ylabel('First Grade')
ax.set_zlabel('Daily Alcohol')

ax.scatter(student_dataset.G1, student_dataset.G3, student_dataset['Dalc'], c='r', marker='.')
plt.show()
```
